### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head(100)

```

#1. Análisis de facturación



```python
ingreso_tienda_1 = int(tienda["Precio"].sum())

ingreso_tienda_2 = int(tienda2["Precio"].sum())
ingreso_tienda_3 = int(tienda3["Precio"].sum())
ingreso_tienda_4 = int(tienda4["Precio"].sum())


facturacion = {
    'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
    'Ingresos': [ingreso_tienda_1, ingreso_tienda_2, ingreso_tienda_3, ingreso_tienda_4]
}

df_facturacion = pd.DataFrame(facturacion, columns=['Tienda', 'Ingresos'])
df_facturacion_sorted = df_facturacion.sort_values(by='Ingresos', ascending=True)
df_facturacion_sorted
```

```python
print(f"Ingreso de la tienda 1: {ingreso_tienda_1}")
print(f"Ingreso de la tienda 2: {ingreso_tienda_2}")
print(f"Ingreso de la tienda 3: {ingreso_tienda_3}")
print(f"Ingreso de la tienda 4: {ingreso_tienda_4}")

```

# 2. Ventas por categoría

```python
ventas_categoria_tienda_1 = tienda.groupby("Categoría del Producto")["Precio"].sum().astype(int)
ventas_categoria_tienda_2 = tienda2.groupby("Categoría del Producto")["Precio"].sum().astype(int)
ventas_categoria_tienda_3 = tienda3.groupby("Categoría del Producto")["Precio"].sum().astype(int)
ventas_categoria_tienda_4 = tienda4.groupby("Categoría del Producto")["Precio"].sum().astype(int)



```

```python
print(f"Venta por categoría tienda 1: \n{ventas_categoria_tienda_1}")
print(f"Venta por categoría tienda 2: \n{ventas_categoria_tienda_2}")
print(f"Venta por categoría tienda 3: \n{ventas_categoria_tienda_3}")
print(f"Venta por categoría tienda 4: \n{ventas_categoria_tienda_4}")
```

# 3. Calificación promedio de la tienda


```python
Calificacion_tienda_1 = round(tienda["Calificación"].mean(), 2)
Calificacion_tienda_2 = round(tienda2["Calificación"].mean(), 2)
Calificacion_tienda_3 = round(tienda3["Calificación"].mean(), 2)
Calificacion_tienda_4 = round(tienda4["Calificación"].mean(), 2)

calificacion = {
    'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
    'Calificacion': [Calificacion_tienda_1, Calificacion_tienda_2, Calificacion_tienda_3, Calificacion_tienda_4]
}
```

```python
print(f"Calificación promedio tienda 1: {Calificacion_tienda_1}")
print(f"Calificación promedio tienda 2: {Calificacion_tienda_2}")
print(f"Calificación promedio tienda 3: {Calificacion_tienda_3}")
print(f"Calificación promedio tienda 4: {Calificacion_tienda_4}")
```

# 4. Productos más y menos vendidos

```python
mayor_menor_venta_tienda_1 = tienda.value_counts("Producto").sort_values(ascending=False)
mayor_menor_venta_tienda_2 = tienda2.value_counts("Producto").sort_values(ascending=False)
mayor_menor_venta_tienda_3 = tienda3.value_counts("Producto").sort_values(ascending=False)
mayor_menor_venta_tienda_4 = tienda4.value_counts("Producto").sort_values(ascending=False)
```

```python
print(f"Productos más vendidos tienda 1: \n{mayor_menor_venta_tienda_1}")
print(f"Productos más vendidos tienda 2: \n{mayor_menor_venta_tienda_2}")
print(f"Productos más vendidos tienda 3: \n{mayor_menor_venta_tienda_3}")
print(f"Productos más vendidos tienda 4: \n{mayor_menor_venta_tienda_4}")

```

# 5. Envío promedio por tienda

```python
mean_tienda_1 = round(tienda["Costo de envío"].mean(),2)
mean_tienda_2 = round(tienda2["Costo de envío"].mean(),2)
mean_tienda_3 = round(tienda3["Costo de envío"].mean(),2)
mean_tienda_4 = round(tienda4["Costo de envío"].mean(),2)

costo_envio = pd.DataFrame({
    'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
    'Costo promedio de envío': [mean_tienda_1, mean_tienda_2, mean_tienda_3, mean_tienda_4]
})
```

```python
print(f"Promedio de costo de envío tienda 1: {mean_tienda_1}")
print(f"Promedio de costo de envío tienda 2: {mean_tienda_2}")
print(f"Promedio de costo de envío tienda 3: {mean_tienda_3}")
print(f"Promedio de costo de envío tienda 4: {mean_tienda_4}")
```
